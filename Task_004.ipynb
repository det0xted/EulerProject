```python
def is_palindrome(n):
    s = str(n)
    return s == s[::-1]

largest = 0

for a in range(100, 1000):
    for b in range(a, 1000):
        product = a * b
        if is_palindrome(product) and product > largest:
            largest = product
            best_a, best_b = a, b

print(f"{best_a} x {best_b} = {largest}")
```
