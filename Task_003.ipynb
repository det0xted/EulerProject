```python
def largest_prime_factor(n):
    largest = 1
    divisor = 2

    while divisor * divisor <= n:
        while n % divisor == 0:
            largest = divisor
            n //= divisor
        divisor += 1

    if n > 1:
        largest = n

    return largest

print(largest_prime_factor(600851475143))
```
